#### Odds (승산)

- 위 그림 (b)에서 합격에 대한 확률을 p라고 했을때 odds는 아래와 같이 표현된다.

$$
\begin{equation} 
Odds = \frac{p}{1-p} \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \
\end{equation}
$$

- 식을 잘 살펴보면, 성공 확률(p)을 실패 확률(1 - p)로 나눈 표현임을 알 수 있다. 즉, odds란 성공 확률이 실패 확률보다 몇 배인지 판단할 수 있는 지표이다. 예를들어 성공 확률이 0.5 (p = 0.5)인 경우 실패 확률 역시 0.5 (1 - p = 1 - 0.5 = 0.5)로 odds는 1이 되며, 만약 성공 확률이 0.8 (p = 0.8)인 경우 실패 확률은 0.2 (1 - p = 1 - 0.8 = 0.2)로 odds는 4가 된다.

- odds를 이진형 변수를 예측하는 회귀 모형에 적용해보면 아래와 같다.

$$
\begin{equation} 
\frac{p}{1-p} = \beta_{0} + \beta_{1}x_{1} + \beta_{2}x_{2} + \cdots + \beta_{p}x_{p} + \epsilon \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \
\end{equation}
$$

- 종속변수가 odds로 표현됨으로써 예측 결과가 odds값 1을 기준으로 1보다 크면 합격, 1보다 작으면 불합격의 형식으로 판단할 수 있을 것이다. 하지만 여기서 딱 한가지 아쉬운 점이 있다. 성공 확률 p는 확률값으로 0과 1 사이의 값으로 표현되므로 모형 좌측의 odds는 결국 0부터 $\infty$까지의 범위를 표현할 수 있는 반면, 우측의 선형 모형은 $-\infty$부터 $\infty$까지의 범위를 표현하기 때문에 좌변과 우변의 범위가 일치되지 않는 것이다.

#### Odds 범위의 확장 (로지스틱 회귀의 완성)

- 그렇다면 odds가 표현할 수 있는 범위를 $-\infty$까지 확장시킬 수 있는 모형이 있다면 위의 모형보다 더 적절할 것이다. 해결 방법은 간단하다. odds에 log를 취하는 것이다.

$$
\begin{equation} 
log\left(\frac{p}{1-p}\right) = \beta_{0} + \beta_{1}x_{1} + \beta_{2}x_{2} + \cdots + \beta_{p}x_{p} + \epsilon \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \
\end{equation}
$$

- log함수는 0부터 $\infty$까지의 값을 정의역(진수)으로 사용하는 함수인데 odds가 표현할 수 있는 값이 딱 log함수의 정의역 범위와 일치한다. 또한 log함수의 치역은 $-\infty$부터 $\infty$까지 표현되어 우리가 원하는 $-\infty$까지 값이 확장될 수 있다.

- 위와같이 odds에 log를 취한 Log Odds를 종속변수로 활용한 모형을 로지스틱 회귀(Logistic Regression)이라고 한다.

- 로지스틱 회귀모형을 p에대해 정리해 아래와 같이 표현하면 이를 시그모이드(sigmoid) 함수라고 하며 이는 결국 우리가 찾던 S자 함수가 된다.

$$
\begin{equation} 
y = \frac{e^{\beta_{0} + \beta_{1}x_{1} + \beta_{2}x_{2} + \cdots + \beta_{p}x_{p} + \epsilon}}{1+e^{\beta_{0} + \beta_{1}x_{1} + \beta_{2}x_{2} + \cdots + \beta_{p}x_{p} + \epsilon}} \ \ \ or \ \ \ y = \frac{1}{1+e^{-(\beta_{0} + \beta_{1}x_{1} + \beta_{2}x_{2} + \cdots + \beta_{p}x_{p} + \epsilon)}}\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \
\end{equation}
$$

- 로지스틱 회귀에서 역시 $\beta$값을 추정해야 하는데 이때 최대 우도 추정법(Maximum liklihood estimation, MLE)을 사용하게 된다. 사실 앞서 선형 회귀에서 $\beta$값을 구할 때 OLS방법을 통해 설명했는데, 선형 회귀에서도 MLE 방식을 통해 $\beta$값을 구할 수 있다.(이는 수리통계학 수준의 내용을 다루므로 간단히 구두로 설명한다.) 단, 선형 회귀와 로지스틱 회귀 $\beta$값 추정에 있어 차이점은 선형 회귀는 $\beta$값에 대한 정확한 해(exact solution)가 존재하는 반면 로지스틱 회귀는 $\beta$ (beta)값에 대한 정확한 해를 찾을 수 없는 것이 차이점이다. 로지스틱 회귀에서는 이를 보안하기 위해 기울기 하강(Gradient Descent)과 같은 방법을 사용한다.(여기서는 아래 그림을 통해 간단히 설명하며, 더 자세한 내용은 '수치해석학' 과목을 통해 접할 수 있다.)

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

# matplotlib 한글 문제
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname = "c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family = font_name)
```

```python
from sklearn.datasets import load_breast_cancer
breast_cancer = load_breast_cancer()
print(breast_cancer.DESCR)
```

```python
# 독립변수와 종속변수 분리
X = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
Y = pd.DataFrame(breast_cancer.target, columns=['target'])
```

```python
X.head()
```

```python
X.shape
```

```python
import statsmodels.api as sm
```

```python
# intercept 추가
X = sm.add_constant(X, has_constant='add')
```

```python
X.head()
```

```python
Y.tail()
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
# 학습 데이터, 테스트 데이터 분리
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.7, test_size=0.3, random_state=1234, stratify=Y)
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
# 로지스틱 회귀 모델 생성
logit_m = sm.Logit(Y_train, X_train)
logit_m_train = logit_m.fit(method='powell') # powell, cg, bfgs, lbfgs
```

```python
logit_m_train.summary()
```

```python
Y_predict = logit_m_train.predict(X_test)
print(Y_predict)
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
# 오차행렬(confusion matirx)를 통해 예측 성능 확인
confusion_matrix(Y_test, Y_predict)
```

- 오류가 발생한다. 무엇이 문제일까? 실제 종속변수의 값(Y_test)은 0, 1로 이루어진 반면 예측된 종속변수의 값(Y_predict)은 0.97, 0.52, 0.04과 같은 0과 1에 사이의 값들로 되어있다. 즉, 각 관측대상의 예측이 확률로 되어있는 것이다.(연속형) 이를 0과 1로 판단해 나눠줘야 하는데 이 때 기준값을 주어야 하며 이를 threshold라고 한다.

```python
# Y_predict를 threshold를 기준으로 0과 1로 나누어주는 함수 생성
def cut_off(y,threshold):
    Y = y.copy() # copy함수를 사용하여 이전의 y값이 변화지 않게 함
    Y[Y>threshold]=1
    Y[Y<=threshold]=0
    return(Y.astype(int))
```

```python
Y_predict_t = cut_off(Y_predict,0.5)
Y_predict_t.head()
```

```python
# 오차행렬(confusion matirx)를 통해 예측 성능 확인
confusion_matrix(Y_test, Y_predict_t)
```

```python
accuracy = accuracy_score(Y_test, Y_predict_t)
precision = precision_score(Y_test, Y_predict_t)
recall = recall_score(Y_test, Y_predict_t)
f1 = f1_score(Y_test, Y_predict_t)
print('Accuracy : {0:.3f}'.format(accuracy))
print('Precision : {0:.3f}'.format(precision))
print('Recall : {0:.3f}'.format(recall))
print('F1 : {0:.3f}'.format(f1))
```

- 근데 과연 threshold를 0.5로 놓는것이 타당한가? 0~1 사이의 값의 변화를 줄때 성능이 어떻게 바뀌는지 확인해보자.

```python
# 임계값에 따른 비교
def perf_eval(cm):
    # Accuracy
    Accuracy = (cm[0, 0] + cm[1, 1]) / sum(cm.reshape(-1, ))
    # Precision
    Precision = cm[1, 1] /sum(cm[:, 1])
    # Recall
    Recall = cm[1, 1] / sum(cm[1])
    # F1-score
    F1 = 2 * Recall * Precision / (Recall + Precision)
    return ([Accuracy, Precision, Recall, F1])

threshold = np.arange(0,1,0.1)
table = pd.DataFrame(columns=['Accuracy','Precision','Recall','F1'])
for i in threshold:
    Y_predict_temp = cut_off(Y_predict,i)
    confusion_matrix_temp = confusion_matrix(Y_test, Y_predict_temp)
    table.loc[i] = perf_eval(confusion_matrix_temp)
table.index.name='threshold'
table.columns.name='performance'
print(table)
```
